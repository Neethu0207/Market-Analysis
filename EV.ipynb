```python
# Create function for unique value Checking
def unique_value_checking(df_column):
    print("List of unique values : " ,df_column.unique())
    print("No of unique values : ", len(df_column.unique()))
```

```python
#importing libraries for visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
unique_columns_counts = []
for i in df.columns:
    unique_columns_counts.append(len(df[i].unique()))
```

```python
# Bar Graph  for unique Value checking
plt.figure(figsize = (12,6))# use plt.figure to create a figure
sns.set_style("white")
bars = plt.bar(df.columns,unique_columns_counts)
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
             str(bar.get_height()), ha='center', va='bottom', fontsize=9)
plt.title("Colums_unique_Values", fontweight = "bold")
plt.xlabel("Columns_Value")
plt.ylabel("Value_Counts")
plt.show()
```

```python
# Iterate through the 'Total Salary' column and print the salary category for each value
for salary_value in df['Total Salary']:
    if salary_value >= 2500000:
        print('high salary')
    elif salary_value >= 1500000:
        print('medium salary')
    else:
        print('low salary')
```

```python
df['Salary_Category'] = df['Total Salary'].apply(lambda salary_value: 'high salary' if salary_value >= 2500000 else ('medium salary' if salary_value >= 1500000 else 'low salary'))
sns.countplot(x='Salary_Category', data=df)

plt.title('Total Salary Category Distribution')
plt.xlabel('Salary Category')
plt.ylabel('Count')
```
